## Data Preproceesing

### column Value

```python
def conv(v):
    if 'M' in v:
        v=v.replace('M','')
        v=float(v[1:])*1000000
        return v
    elif 'K' in v:
        v=v.replace('K','')
        v=float(v[1:])*1000
        return v
```

```python
df1['Value']=df1['Value'].map(conv)
```

### column Wage

```python
def conve(w):
    if 'M' in w:
        w=w.replace('M','')
        w=float(w[1:])*1000000
        return w
    elif 'K' in w:
        w=w.replace('K','')
        w=float(w[1:])*1000
        return w
```

```python
df1['Wage']=df1['Wage'].map(conve)
```

### column Joined

```python
df1['Joined']=df1['Joined'].replace(np.nan,0)
```

```python
def conv_year(j):
    j=int(j[-4:])
    return j
```

```python
df1['Joined']=df1['Joined'].astype('str')
```

```python
df1['Joined']=df1['Joined'].map(conv_year)
```

### column Contract valid until

```python
import datetime
```

```python
df1['Contract Valid Until']=df1['Contract Valid Until'].replace(np.nan,0)
```

```python
def conv_datetime(c):
    if c!='0':
        if len(c)==4:
            c=datetime.datetime.strptime(c,'%Y')
            return c
        else:
            c=c.replace(',','')
            c=datetime.datetime.strptime(c,'%b %d %Y')
            return c
```

```python
df1['Contract Valid Until']=df1['Contract Valid Until'].astype('str')
```

```python
df1['Contract Valid Until']=df1['Contract Valid Until'].map(conv_datetime)
```

### column Height

```python
df1['Height']=df1['Height'].replace(np.nan,0)
```

```python
def conv_height(h):
    if h!='0':
        h=float(h[2:])*(1/12)+float(h[0])
        return h
```

```python
df1['Height']=df1['Height'].astype('str')
```

```python
df1['Height']=df1['Height'].map(conv_height)
```

### column Weight

```python
df1['Weight']=df1['Weight'].replace(np.nan,0)
```

```python
def conv_weight(w):
    if 'lbs' in w:
        w=w.replace('lbs','')
        w=float(w[0:])
        return w
```

```python
df1['Weight']=df1['Weight'].astype('str')
```

```python
df1['Weight']=df1.Weight.map(conv_weight)
```

### column Release Clause

```python
df1['Release Clause']=df1['Release Clause'].replace(np.nan,0)
```

```python
def conv_release_clause(rc):
    if 'M' in rc:
        rc=rc.replace('M','')
        rc=float(rc[1:])*1000000
        return rc
    elif 'K' in rc:
        rc=rc.replace('K','')
        rc=float(rc[1:])*1000
        return rc
```

```python
df1['Release Clause']=df1['Release Clause'].astype('str')
```

```python
df1['Release Clause']=df1['Release Clause'].map(conv_release_clause)
```
